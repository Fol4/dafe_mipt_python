#pd.Series

Тип данных pd.Series представляет собой одномерный набор данных. Отсутствующий данные записываются как np.nan. Они не участвуют в вычислении средних, среднеквадратичных отклонений и т.д.
Создание

Создадим Series из списка температур


```python
temperature = [0, 0, 1, np.nan, -1, -3, -5, -3, -6, np.nan, -8,- 9, -11, -14]
ser_1 = pd.Series(temperature)
ser_1
```

```python
data = ['2021-01-' + str(i).zfill(2) for i in range(1, len(temperature) + 1)]
ser_2 = pd.Series(temperature, data)
ser_2
```

```python
ser_2.isnull()
```

```python
ser_2.dropna(how="any")
```

```python
ser_2.fillna(value=-7)
```

```python
ser_3 = pd.Series(temperature, data, name='Temperature')
ser_3
```

##Индексирование

```python
print(ser_3[0])

print('-----------')

print(ser_3[1:3])

print('-----------')

print(ser_3[::-1])
```

#DataFrame

Объект DataFrame лучше всего представлять себе в виде обычной таблицы и это правильно, ведь DataFrame является табличной структурой данных. В любой таблице всегда присутствуют строки и столбцы. Столбцами в объекте DataFrame выступают объекты Series, строки которых являются их непосредственными элементами.

```python
df1 = pd.DataFrame({
     'country': ['Kazakhstan', 'Russia', 'Belarus', 'China'],
     'population': [17.04, 143.5, 9.5, 1412],
     'square': [2724902, 17125191, 207600, 9597000]})
df1
```

```python
data = [['Kazakhstan', 17.04, 2724902], ['Russia', 143.5, 17125191], ['Belarus', 9.5, 207600], ['China', 1412, 9597000]]
df2 = pd.DataFrame(data, columns=['country', 'population', 'square'])
df2
```

```python
df2.columns
```

```python
df2.values
```

```python
df2.columns = ['Страна', 'Население', 'Плошадь']
df2
```

```python
df1.index = ['KZ', 'RU', 'BY', 'CH']
df1
```

```python
df1.index.name = 'Country Code'
df1
```

###Индиксация по строкам при помощи .loc и .iloc

.loc и .iloc --- это два взаимозаменяемых атрибута, которые позволяют индексировать по обеим осям сразу. Путаницы не возникает из-за фиксированного порядка перечисления осей.



1.  .loc - используется для доступа по строковой метке
2.  .iloc - используется для доступа по числовому значению (начиная от 0)



```python
df1.loc['KZ']
```

```python
df1.iloc[0]
```

```python
df1.loc[['KZ', 'RU'], 'population']
```

```python
df1.loc['KZ' : 'BY', :]
```

```python
df1[df1.population > 10][['country', 'square']]
```

```python
left = pd.DataFrame({"key": ["foo", "foo"], "lval": [1, 2]})
right = pd.DataFrame({"key": ["foo", "foo"], "rval": [4, 5]})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, on="key")
```

```python
left = pd.DataFrame({"key": ["foo", "bar"], "lval": [1, 2]})
right = pd.DataFrame({"key": ["foo", "bar"], "rval": [4, 5]})
```

```python
left
```

```python
right
```

```python
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'], 'value': [1, 2, 3, 5]})
df1
```

```python
df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'], 'value': [5, 6, 7, 8]})
df2
```

```python
df1.merge(df2, left_on='lkey', right_on='rkey')
```

```python
df1.merge(df2, left_on='lkey', right_on='rkey', suffixes=('_left', '_right'))
```
